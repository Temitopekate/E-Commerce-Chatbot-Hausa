```python
import json
import matplotlib.pyplot as plt

# 1. Load the intent data from intents.json
with open("intents.json", "r") as file:
    intent_data = json.load(file)

# 2. Extract messages
messages = {}

for intent in intent_data["intents"]:
    tag = intent["tag"]
    patterns = intent["patterns"]
    responses = intent["responses"]
    
    messages[tag] = patterns + responses

# 3. Plot the distribution
tags = list(messages.keys())
counts = [len(messages[tag]) for tag in tags]

plt.figure(figsize=(12, 6))
plt.barh(tags, counts, color='skyblue')
plt.xlabel('Number of Messages')
plt.ylabel('Tags')
plt.title('Distribution of Messages by Tags')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# 1. Load the intent data from intents.json
with open("intents.json", "r") as file:
    intent_data = json.load(file)

# 2. Extract messages
messages = {}

for intent in intent_data["intents"]:
    tag = intent["tag"]
    patterns = intent["patterns"]
    responses = intent["responses"]
    
    messages[tag] = len(patterns) + len(responses)

# 3. Plot the pie chart
labels = list(messages.keys())
sizes = list(messages.values())

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.tab20.colors)
plt.title('Distribution of Messages by Tags')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
